# Univariate Analysis

```python
# Histogram
plt.figure(figsize=(10, 6))
sns.barplot(train['sub-category'])
plt.title('Distribution of appliances brought')
plt.xlabel('Count')
plt.ylabel('Appliances')
plt.show()
```

As can be seen, the purchase and shipping of most tof the items are roughly the same, however, the higher count lies for copiers, closely followed by supplies, bookcases and chairs.

```python
plt.ticklabel_format(style='plain')

sns.histplot(train['sales'], kde=True)
plt.xlabel('price')
plt.title('Histogram Plot')
plt.show()
```

```python
plt.figure(figsize=(30, 6))
sns.countplot(x='state', data=train, palette='pastel')
plt.title('Count of purchase per state')
plt.xlabel('States')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

With a major count of numbers, California, as can be seen, has the higest count of purchase per state.

```python
plt.figure(figsize=(11,6))
sns.countplot(data=train, x='region', alpha=.6, color="blue")
plt.title('Purchase and shipping across regions',fontsize=18)
plt.xlabel('Region',fontsize=16)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.ylabel('');
```

 The Western regions stand out prominently, boasting the highest count of purchases when juxtaposed with other geographic areas. This trend underscores a notable preference or heightened economic activity in the Western territories, indicating a robust and thriving marketplace.

```python
# Boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(x=train['sales'])
plt.title('Boxplot of Sales')
plt.xlabel('Sale')
plt.show()
```

```python

data = [2931, 1007, 883]
keys = ['Office Supplies', 'Furniture', 'Technology']

# define Seaborn color palette to use
palette_color = sns.color_palette('pastel')

# plotting data on chart
plt.pie(data, labels=keys, colors=palette_color, autopct='%.0f%%')

# displaying chart
plt.show()
```

 Office supplies, serving as the backbone of day-to-day operations, constitute a substantial share of 61 percent. Furniture, a pivotal aspect of workspace aesthetics and functionality, commands a significant allocation of 21 percent. Technology, a driving force behind modern business operations, holds a noteworthy distribution share of 18 percent.

```python
data = [2571, 1359, 891]
keys = ['Consumer', 'Corporate', 'Home Office']

# declaring exploding pie
explode = [0.05] * len(data)
# define Seaborn color palette to use
palette_color = sns.color_palette('pastel')

# plotting data on chart
plt.pie(data, labels=keys, colors=palette_color, explode=explode, autopct='%.0f%%')

# displaying chart
plt.show()
```

Consumer transactions take the lead, constituting 53 percent of the overall distribution, indicative of the substantial influence and purchasing power wielded by individual buyers in the market. Meanwhile, corporate transactions hold a significant share at 28 percent, underscoring the pivotal role that businesses play in the economic landscape. The home-office sector follows closely behind, contributing 18 percent to the distribution.

```python
# Load the data
df = pd.read_csv('/content/train.csv')

# Calculate the correlation matrix
correlation_matrix = df.corr()

# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix')
plt.show()
```

The above correlation matrix shows the correlation between pairs of numeric variables. The values range from 0 to 1, where 1 indicates a perfect positive correlation, 0 indicates no correlation. Positive values indicate a positive relationship, while negative values indicate a negative relationship. Variables with a higher absolute correlation value are better correlated.

```python
plt.figure(figsize=(20, 15))
sns.scatterplot(x='sales', y='sub-category', data=train, hue='state', palette='viridis', s=300)
plt.title('Scatter Plot: Price vs Cylinders')
plt.xlabel('Sales')
plt.ylabel('Sub-category')
plt.legend(title='Company')
plt.show()
```

```python
plt.figure(figsize=(28,8))
sns.lineplot(x='sales', y='sub-category', hue='segment', data=train, palette='Set2')
plt.title('Scatter Chart: Price by Company with Body Type')
plt.xlabel('Company')
plt.ylabel('Price')
plt.legend(title='Body Type')
plt.show()
```
